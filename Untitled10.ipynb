```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from webdriver_manager.chrome import ChromeDriverManager

options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-blink-features=AutomationControlled')
options.add_experimental_option('excludeSwitches', ['enable-automation'])

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

driver.get("https://www.arabam.com/ikinci-el/arazi-suv-pick-up-istanbul-otomatik-sahibinden?fuel=Benzin&fuel=Dizel&fuel=Elektrik&fuel=Hibrit&fuel=LPG&bodytype=Crossover&status=%C4%B0kinci+El&damagestatus=Boyas%C4%B1z%2C+De%C4%9Fi%C5%9Fensiz+ve+Tramersiz&cekis=Arkadan+%C4%B0ti%C5%9F&cekis=%C3%96nden+%C3%87eki%C5%9F&cekis=4WD+%28S%C3%BCrekli%29&cekis=AWD+%28Elektronik%29&gear=D%C3%BCz&gear=Yar%C4%B1+Otomatik")


try:
    cookie_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "Kabul Et") or contains(text(), "Accept")]'))
    )
    cookie_button.click()
    print("Çerezler kabul edildi.")
except:
    print("Çerez onay butonu bulunamadı veya zaten kabul edildi.")

links = []


page_number = 1

while True:
    print(f"{page_number}. sayfa işleniyor...")

    ilanlar = driver.find_elements(By.XPATH, '//tr[contains(@id, "listing")]/td[2]/a')
    for ilan in ilanlar:
        link = ilan.get_attribute('href')
        if link and link not in links:
            links.append(link)
            print(f"Link bulundu: {link}")

    try:
       
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        
        
        next_page_button = driver.find_element(By.XPATH, f'//a[@id="paging_{page_number + 1}"]')
        ActionChains(driver).move_to_element(next_page_button).click().perform()
        page_number += 1
        time.sleep(3)
    
    except:
        print("Son sayfaya ulaşıldı veya sayfa değiştirilemedi.")
        break


if links:
    with open('arabam2_links.csv', 'w', newline='', encoding='utf-8-sig') as file:
        writer = csv.writer(file)
        writer.writerow(["Link"])
        for link in links:
            writer.writerow([link])
    print(f"Toplam {len(links)} ilan linki başarıyla kaydedildi")
else:
    print("Hiç ilan linki bulunamadı!")

driver.quit()
```

```python
import csv
import time
import random
import logging
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException, TimeoutException

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

CONFIG = {
    'timeout': 15,  
    'wait_times': (1, 3), 
}

def create_driver():
    """Chrome WebDriver oluşturur."""
    options = webdriver.ChromeOptions()
    options.add_argument('--disable-blink-features=AutomationControlled')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--ignore-certificate-errors')
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    return driver

def accept_cookies(driver):
    """Çerezleri kabul eder."""
    try:
        cookie_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "Kabul") or contains(text(), "Accept")]'))
        )
        cookie_button.click()
        logging.info("Çerezler kabul edildi")
    except Exception:
        logging.info("Çerez butonu bulunamadı veya zaten kabul edilmiş")

def fetch_listing_details(driver, link):
    """İlan detaylarını çeker."""
    details = {'link': link}
    try:
        driver.get(link)
        WebDriverWait(driver, CONFIG['timeout']).until(
            EC.presence_of_element_located((By.TAG_NAME, 'body'))
        )
        accept_cookies(driver)

        locators = {
            'marka': '//div[contains(text(), "Marka")]/following-sibling::div',
            'model': '//div[contains(text(), "Seri")]/following-sibling::div',
            'yıl': '//div[contains(text(), "Yıl")]/following-sibling::div',
            'km': '//div[contains(text(), "Kilometre")]/following-sibling::div',
            'kasa_tipi': '//div[contains(text(), "Kasa Tipi")]/following-sibling::div',
            'motor_gücü': '//div[contains(text(), "Motor Gücü")]/following-sibling::div',
            'çekiş': '//div[contains(text(), "Çekiş")]/following-sibling::div',
            'renk': '//div[contains(text(), "Renk")]/following-sibling::div',
            'garanti': '//div[contains(text(), "Garanti")]/following-sibling::div',
            'kapı_sayısı': '//div[contains(text(), "Kapı Sayısı")]/following-sibling::div',
            'takas_durumu': '//div[contains(text(), "Takas")]/following-sibling::div',
            'fiyat': '//div[contains(@class, "price") or contains(text(), "Fiyat")]/following-sibling::div' 
        }
        
        for key, xpath in locators.items():
            try:
                element = driver.find_element(By.XPATH, xpath)
                details[key] = element.text.strip()
            except NoSuchElementException:
                details[key] = 'Bulunamadı'

        logging.info(f"İlan çekildi: {details}")
        time.sleep(random.uniform(*CONFIG['wait_times']))  

    except TimeoutException:
        logging.error(f"İlan yüklenemedi: {link}")
        details['error'] = "Zaman aşımı"
    except Exception as e:
        logging.error(f"Hata: {link} - {str(e)}")
        details['error'] = str(e)

    return details

def save_to_csv(data, filename='arabam_car_details2.csv'):
    """Veriyi CSV'ye ekler."""
    file_exists = False
    try:
        file_exists = open(filename).readline()
    except FileNotFoundError:
        pass
    
    with open(filename, 'a', newline='', encoding='utf-8-sig') as f:
        writer = csv.DictWriter(f, fieldnames=data.keys())
        if not file_exists:
            writer.writeheader()  
        writer.writerow(data)

def main():
    """Ana iş akışı."""
    try:
        with open('arabam2_links.csv', 'r', encoding='utf-8') as f:
            links = [line.strip() for line in f.readlines()]
        logging.info(f"{len(links)} ilan linki bulundu")
    except Exception as e:
        logging.error(f"CSV dosyası okuma hatası: {str(e)}")
        return

    driver = create_driver()

    try:
        for i, link in enumerate(links):
            result = fetch_listing_details(driver, link)
            save_to_csv(result)  
            logging.info(f"{i+1}/{len(links)} işlendi: {link}")
    finally:
        driver.quit()

    logging.info("Tüm veriler kaydedildi.")

if __name__ == "__main__":
    main()
```
